```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gjør samme som forrige oppgave, men med ny formel

def f(x):
    return np.exp(x)

x = 1.5
f_exact = f(x)

h_values = [10**(-i) for i in range(1, 17)]

results = []

for h in h_values:
    f_approx = (f(x + h) - f(x - h)) / (2 * h)
    error = abs(f_exact - f_approx)
    results.append((h, f_approx, error))

df = pd.DataFrame(results, columns=["h-values", "f'(x) approx", "abs(error)"])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# semilogx
ax1.semilogx(df["h-values"], df["abs(error)"], marker='o')
ax1.set_xlabel("h-values")
ax1.set_ylabel("abs(error)")
ax1.set_title("Semi-logx")
ax1.grid(True, which="both", ls="--")

# log-log
ax2.loglog(df["h-values"], df["abs(error)"], marker='o')
ax2.set_xlabel("h-values")
ax2.set_ylabel("abs(error)")
ax2.set_title("Log-log")
ax2.grid(True, which="both", ls="--")

plt.tight_layout()
plt.show()

print('Som vi ser av grafen under blir resultatet mer nøyaktig enn forrige oppgave, og det går åt skogen når h blir lik 10^-16')

df
```
