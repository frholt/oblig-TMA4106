```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Likt som de to forrige oppgavene, men med en enda mer nøyaktig formel :) 

def f(x):
    return np.exp(x)

x0 = 1.5
f_exact = f(x0)

def formel(x, h):
    return (f(x-2*h) - 8*f(x-h) + 8*f(x+h) - f(x+2*h)) / (12*h)

h_values = [10**(-i) for i in range(1, 17)]
results = []

for h in h_values:
    f_approx = formel(x0,h)
    error = abs(f_approx - f_exact)
    results.append((h, f_approx, error))
    
df = pd.DataFrame(results, columns=["h-values", "f'(x) approx", "abs(error)"])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# semilogx
ax1.semilogx(df["h-values"], df["abs(error)"], marker='o')
ax1.set_xlabel("h-values")
ax1.set_ylabel("abs(error)")
ax1.set_title("Semi-log")
ax1.grid(True, which="both", ls="--")

# log-log
ax2.loglog(df["h-values"], df["abs(error)"], marker='o')
ax2.set_xlabel("h-values")
ax2.set_ylabel("abs(error)")
ax2.set_title("Log-log")
ax2.grid(True, which="both", ls="--")

plt.tight_layout()
plt.show()

df
```
